```python
import pandas as pd
import numpy as np
```

```python
teams = pd.read_csv("teams.csv")
```

```python
teams
```

```python
X = teams[["athletes","prev_medals"]].copy()
Y = teams[["medals"]].copy()
```

making two matrices

```python
X
```

```python
Y
```

```python
X["intercept"] = 1
```

```python
X
```

```python
X = X[["intercept", "athletes","prev_medals"]]
```

```python
X_T = X.T
```

T is for transpose.

```python
X_T
```

```python
B = np.linalg.inv(X_T @ X) @ X_T @ Y
```

B is formula for calculating coefficients.

@ is matrix multiplication operator, we are multiplying x transpose with x and taking inverse of the whole matrix.

```python
B
```

```python
B.index = X.columns
```

```python
B
```

```python
predictions = X @ B
```

```python
predictions
```

```python
SSR = ((Y - predictions) ** 2).sum()
```

```python
SST = ((Y - Y.mean()) ** 2).sum()
```

```python
SSR
```

```python
SST
```

```python
R2 = 1 - (SSR/SST)
```

```python
R2
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr.fit(teams[["athletes","prev_medals"]],teams[["medals"]])
```
